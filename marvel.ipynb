```python
import pandas as pd
from pandas import DataFrame, read_csv
```

```python
porgat = pd.read_csv('./porgat.txt', sep='\t')
porgat = porgat.set_index('ID')
porgat
```

```python
vert1 = pd.read_csv('./vert1.txt', sep='\t', header=False)
vert1 = vert1.set_index('ID')
vert2 = pd.read_csv('./vert2.txt', sep='\t', header=False)
vert2 = vert2.set_index('ID')

verts = pd.concat([vert1, vert2])
verts
```

```python
edges = pd.read_csv('./edges.txt', sep='\t', header=False)
edges = edges.set_index('ID')
edge_list = edges['EDGES'].apply(lambda x: x.split())
edge_list = edge_list.groupby(edges.index).sum()
```

```python
edge_frame = edge_list.to_frame()
edge_frame.columns = ['EDGES']
edge_frame
```

```python
df = pd.merge(edge_frame, porgat, how='left', left_index=True, right_index=True)
df = pd.merge(df, verts, how='left', left_index=True, right_index=True)
df
```
